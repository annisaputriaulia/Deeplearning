# **Sequence Model**

Sequence Model merupakan model yang menerima input berbentuk sequence dan mengeluarkan output sequence juga. Terdapat beberapa Sequence Model, tapi yang terpopuler saat ini ialah RNN

**Autoregressive Models**



Autoregressive (AR) Models merupakan model feed-forward yang memprediksi nilai dari output t (sekarang) berdasarkan nilai masa lalu (t-1). AR models yang memprediksi nilai titik sekarang berdasarkan 1 titik di belakangnya (atau 1 lags) disebut AR(1), jika memprediksi berdasarkan 2 titik disebut AR(2), dst.

**Markov Models**

Markove Models merupakan metode stokastik untuk sistem yang berubah secara acak. Berbeda dengan AR model, Markov Model mengasumsikan keadaan/state masa depan tidak bergantung pada keadaan masa lalu. Model Markov ini mampu menunjukkan probabilitas semua state yang mungkin terjadi.

**Recurrent Neural Network (RNN)**



RNN merupakan model sequence yang memperkenalkan variabel status/state untuk menyimpan informasi di titik-titik data sebelumnya, bersama dengan input saat ini untuk menentukan output saat ini.
